```python
%matplotlib inline
```

```python
# Import dependencies 
import matplotlib.pyplot as plt
```

```python
# Set the x axis to a list of string for each month 
x_axis = ["Jan", "Feb","Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month 
y_axis =[10.02,23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot 
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot. 
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#Create the plot and add a label for the legend 
plt.plot(x_axis, y_axis, label = "Boston")

#Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45. 
plt.ylim(0,45)

# Create a title 
plt.title("PyBer Fare by Month")

#Add the legend. 
plt.legend()
```

```python
# Create the plot. 
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = "Boston")

# Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45.
plt.ylim(0,45)

#Create a title 
plt.title("PyBer Fare by Month")

# Add a grid 
plt.grid()

# Add the legend. 
plt.legend()
```

```python
# Skill Drill in 5.1.4
ax = plt.axes()
ax.plot(x_axis, y_axis, color = "green", marker = "*", linewidth = 2, label = "Boston")
ax.legend()
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,45)
ax.grid()
```

```python
# Create the bar plot 
plt.bar(x_axis, y_axis, color = "green", label = "Boston")

#Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create a title 
plt.title("PyBer Fare by Month")

#Add the legend 
plt.legend()
```

```python
# Create the plot 
plt.barh(x_axis, y_axis, color = "magenta", )
plt.gca().invert_yaxis()
```

```python
# Skill Drill 5.1.5 
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.xlabel = ("Fare($)")
plt.ylabel = ("Date")
plt.legend()
plt.title ("PyBer Fare by Month")
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month 
colors =["slateblue","magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
